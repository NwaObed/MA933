# SRM with absorbing boundary

```python
p = q = 0.5
S = list(range(-5,5))

tmax = 100
Z = []
sim = 500
tA = np.zeros(sim) #absorption time

for j in range(sim):
    X0 = 0 #initial point
    t = [] #time when walker enters absorption
    walker = X0
    #print(t)
    for i in range(tmax):
        r = np.random.rand()
        #Z.append(walker)
        if walker == -5 or walker==5:
            walker = walker #gets absorbed without moving
            t.append(i)

        elif r >= p:
            walker += 1 #move one step forward

        else:
            walker -= 1 #move one step backward
    try:
        tA[j] = min(t)
    except ValueError:
        continue
print(0 in tA)
tA 
```
