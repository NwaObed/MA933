```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def wright_fisher(L,T):
    
    X0 = np.arange(1,L+1)
    MC = np.zeros((T,L),dtype=int)
    MC[0] = X0
    for i in range(1,T):
        current_state = MC[i-1,:]
        sample = np.random.randint(current_state[0],current_state[-1],L)
        MC[i] = np.sort(sample)
    
    return MC

M = wright_fisher(L,T)
L = 10 #population size
T = 5
plt.figure(figsize=(15,10))
sns.heatmap(M)
```
